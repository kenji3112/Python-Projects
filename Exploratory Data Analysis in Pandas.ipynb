```python
# Select only the numeric columns
numeric_df = df.select_dtypes(include = [np.number])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

corr_matrix
```

```python
sns.heatmap(numeric_df.corr(), annot= True, cmap='coolwarm', fmt='.2f', annot_kws={"size": 10})
plt.rcParams['figure.figsize'] = (20,7)
plt.show()
```

```python
continent = df.groupby('Continent')

```

```python
numeric_columns = df.select_dtypes(include='number').columns
```

```python
mean_values = continent[numeric_columns].mean()

mean_values
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
mean_values = continent[['1970 Population', '1980 Population', '1990 Population',
       '2000 Population', '2010 Population', '2015 Population',
       '2020 Population', '2022 Population']].mean().sort_values(by = "2022 Population", ascending = False)

df2 = mean_values 
df2
```
